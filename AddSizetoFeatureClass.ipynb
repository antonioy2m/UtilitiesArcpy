### Code to add the size category to rural and urban parcels


*Usage:* 

```
fc = r'path/to/feature/class'

sizeCat(fc,'U') # 'U' to use Urban dictionary, empty to use Rural

```

````python
import pandas

#Dictionary with the parcel size category in ha
sizeRur_dict = {
    "A. Hasta 0.5 ha": (0, 0.5),
    "B. Mayor a 0.5 ha - hasta 1 ha": (0.5, 1),
    "C. Mayor a 1 ha - hasta 2.5 ha": (1, 2.5),
    "D. Mayor a 2.5 ha - hasta 3 ha": (2.5, 3),
    "E. Mayor a 3 ha - hasta 5 ha": (3, 5),
    "F. Mayor a 5 ha - hasta 10 ha": (5, 10),
    "G. Mayor a 10 ha - hasta 20 ha": (10, 20),
    "H. Mayor a 20 ha - hasta 50 ha": (20, 50),
    "I. Mayor a 50 ha - hasta 100 ha": (50, 100),
    "J. Mayor a 100 ha - hasta 200 ha": (100, 200),
    "K. Mayor a 200 ha - hasta 500 ha": (200, 500),
    "L. Mayor a 500 ha - hasta 1000 ha": (500, 1000),
    "M. Mayor a 1000 ha - hasta 2000 ha": (1000, 2000),
    "N. Mayor a 2000 ha - hasta 5000 ha": (2000, 5000),
    "O. Mayor a 5000 ha - hasta 10000 ha": (5000, 10000),
    "P. Mayor a 10000 ha": (10000, float("inf"))
}

#Dictionary with the parcel size category in square meters
sizeUrb_dict = {
    "A. Hasta 50 m²": (0, 50),
    "B. Mayor a 50 m² - hasta 100 m²": (50, 100),
    "C. Mayor a 100 m² - hasta 200 m²": (100, 200),
    "D. Mayor a 200 m² - hasta 500 m²": (200, 500),
    "E. Mayor a 500 m² - hasta 1.000 m²": (500, 1000),
    "F. Mayor a 1.000 m² - hasta 2.000 m²": (1000, 2000),
    "G. Mayor a 2.000 m² - hasta 5.000 m²": (2000, 5000),
    "H. Mayor a 5.000 m² - hasta 10.000 m²": (5000, 10000),
    "I. Mayor a 10000 m²": (10000, float("inf"))
}

# *!*!*!*!*!*!*!*!*!*!*!*!*!*!*!*!*!*!*!*!*!*!*!*!*!*!*!*!*!*!*!*!*!*!*!*!*!*!*!*!*!*!*!*!*!*!*!*!*!*!*!*!*!*! #

#Function to add fields and calculate the range
def sizeCat(fc, urbRural='R'):
    #add the fields
    arcpy.management.AddField(fc,'rangoTamanio','TEXT')
    if urbRural == 'U':
        arcpy.management.AddField(fc,'aream2','DOUBLE')
        dicti = sizeUrb_dict
        with arcpy.da.UpdateCursor(fc, ['Aream2', 'SHAPE@', "rangoTamanio"]) as cursor:
            # For each row, calculate the area
            for row in cursor:
                area = row[1].getArea('PLANAR','SQUAREMETERS')
                row[0] = area
                # for each row calculate the category and break when find it
                for categoria, rango in dicti.items():
                    if rango[0] < area <= rango[1]:
                        row[2] = categoria
                        break
                # Update the cursor with the updated value
                cursor.updateRow(row)
    else:
        arcpy.management.AddField(fc,'Areaha','DOUBLE')
        dicti = sizeRur_dict
        with arcpy.da.UpdateCursor(fc, ['Areaha', 'SHAPE@', "rangoTamanio"]) as cursor:
        # For each row, calculate the area
            for row in cursor:
                area = row[1].getArea('PLANAR','HECTARES')
                row[0] = area
                # for each row calculate the category and break when find it
                for categoria, rango in dicti.items():
                    if rango[0] < area <= rango[1]:
                        row[2] = categoria
                        break
                # Update the cursor with the updated value
                cursor.updateRow(row)
````
